12. Псевдообращение Мура-Пенроуза применяется в [машинах экстремального обучения](https://link.springer.com/article/10.1007/s11042-021-11007-7) - современных искусственных нейронных сетях, которые не требуется обучать с помощью классических градиентных методов, что существенно ускоряет процесс обучения. Предлагается реализовать обращение Мура-Пенроуза в виде функции `moore_penrose` для матрицы $\mathbf{A}$ по формуле:

$
\mathbf{H}^† = (\mathbf{H}^T \mathbf{H})^{-1}\mathbf{H}^T
$.

```python
def add(matrix_a, matrix_b):
    """1. Сложение двух матриц одинакового размера."""
    rows, cols = len(matrix_a), len(matrix_a[0])
    result = [[0.0 for _ in range(cols)] for _ in range(rows)]
    for i in range(rows):
        for j in range(cols):
            result[i][j] = float(matrix_a[i][j] + matrix_b[i][j])
    return result


def multiply(matrix_a, matrix_b):
    """2. Поэлементное произведение матриц (Произведение Адамара)."""
    rows, cols = len(matrix_a), len(matrix_a[0])
    result = [[0.0 for _ in range(cols)] for _ in range(rows)]
    for i in range(rows):
        for j in range(cols):
            result[i][j] = float(matrix_a[i][j] * matrix_b[i][j])
    return result


def dot(matrix_a, matrix_b):
    """3. Классическое матричное произведение (строка на столбец)."""
    rows_a, cols_a = len(matrix_a), len(matrix_a[0])
    cols_b = len(matrix_b[0])
    result = [[0.0 for _ in range(cols_b)] for _ in range(rows_a)]
    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):
                result[i][j] += float(matrix_a[i][k] * matrix_b[k][j])
    return result


def transpose(matrix_a):
    """4. Транспонирование матрицы."""
    rows, cols = len(matrix_a), len(matrix_a[0])
    result = [[0.0 for _ in range(rows)] for _ in range(cols)]
    for i in range(cols):
        for j in range(rows):
            result[i][j] = float(matrix_a[j][i])
    return result


def submatrix(matrix_a, row_to_exclude, col_to_exclude):
    """5. Вычеркивание row_to_exclude строки и col_to_exclude столбца."""
    result = []
    for i, row in enumerate(matrix_a):
        if i == row_to_exclude:
            continue
        new_row = [float(elem) for j, elem in enumerate(row) if j != col_to_exclude]
        result.append(new_row)
    return result


def smalldet(matrix_a):
    """6. Вычисление определителя матрицы 2x2."""
    return float(matrix_a[0][0] * matrix_a[1][1] - matrix_a[0][1] * matrix_a[1][0])


def det(matrix_a, i=0):
    """7. Рекуррентное вычисление определителя матрицы n x n."""
    n = len(matrix_a)
    # КРИТИЧЕСКОЕ ИСПРАВЛЕНИЕ: Базовый случай для матриц 1х1 (важно для инверсии 2х2)
    if n == 1:
        if isinstance(matrix_a[0], list):
            return float(matrix_a[0][0])
        return float(matrix_a[0])

    if n == 2:
        return smalldet(matrix_a)

    total_det = 0.0
    for j in range(n):
        a_ij = matrix_a[i][j]
        sign = (-1) ** (i + j)
        sub_A = submatrix(matrix_a, i, j)
        total_det += sign * a_ij * det(sub_A, i=0)
    return float(total_det)


def minor(matrix_a, i, j):
    """8. Вычисление дополнительного минора элемента i, j."""
    return det(submatrix(matrix_a, i, j), i=0)


def alg(matrix_a, i, j):
    """9. Вычисление алгебраического дополнения элемента i, j."""
    return float(((-1) ** (i + j)) * minor(matrix_a, i, j))


def algmatrix(matrix_a):
    """10. Построение матрицы алгебраических дополнений."""
    n = len(matrix_a)
    result = [[0.0 for _ in range(n)] for _ in range(n)]
    for row_idx in range(n):
        for col_idx in range(n):
            result[row_idx][col_idx] = alg(matrix_a, row_idx, col_idx)
    return result


def inv(matrix_a):
    """11. Вычисление классической обратной матрицы."""
    determinant = det(matrix_a, i=0)
    if determinant == 0:
        raise ValueError("Обратной матрицы не существует (det = 0).")

    adj_matrix = transpose(algmatrix(matrix_a))

    n = len(matrix_a)
    result = [[0.0 for _ in range(n)] for _ in range(n)]
    for row_idx in range(n):
        for col_idx in range(n):
            result[row_idx][col_idx] = float(adj_matrix[row_idx][col_idx] / determinant)
    return result


def moore_penrose(matrix_h):
    """12. Вычисление псевдообращения Мура - Пенроуза."""
    matrix_h_t = transpose(matrix_h)
    inner_product = dot(matrix_h_t, matrix_h)

    try:
        inner_inverse = inv(inner_product)
    except ValueError:
        raise ValueError("Матрица (H^T * H) вырождена, псевдообращение невозможно.")

    return dot(inner_inverse, matrix_h_t)


def print_matrix(matrix, precision=4):
    """Вспомогательная функция для форматированного вывода результатов."""
    for row in matrix:
        formatted_row = [round(float(x), precision) if isinstance(x, (int, float)) else x for x in row]
        print(formatted_row)



if __name__ == "__main__":
    print("--- Пункт 12: Псевдообращение Мура - Пенроуза moore_penrose(H) ---")
    H_12 = [[1, 2], [3, 4], [5, 6]]
    print_matrix(moore_penrose(H_12))
```
